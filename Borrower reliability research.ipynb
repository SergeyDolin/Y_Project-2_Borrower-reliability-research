```python
df['price_for_sq_m'] = (df['last_price'] / df['total_area']).round()
df.head(10)
```

### Добавляем столбцы с днем недели, месяцем и годом публикации объявления

Для того, чтобы обеспечить возможность анализа данных по годам, месяцам и дням недели, добавим в датасет соответствующие столбцы. День недели публикации объявления выразим числом от 0 до 6 (0 — понедельник, 1 — вторник и так далее).

```python
df['week_day_exposition'] = df['first_day_exposition'].dt.weekday
df['month_exposition'] = df['first_day_exposition'].dt.month
df['year_exposition'] = df['first_day_exposition'].dt.year
df.head(10)
```

### Добавляем столбец с типом этажа квартиры

Добавим в датасет столбец с типом этажа квартиры (значения — «первый», «последний», «другой»).

```python
def group_floors(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

df['group_floors'] = df.apply(group_floors, axis=1)
df.head(10)
```

### Добавляем столбец с расстоянием до центра в км

Переведем расстояние до центра в км и округлим до целых значений, данные сохраним в отдельном столбце.

```python
df['cityCenters_nearest_km'] = (df['cityCenters_nearest'] / 1000).round()
df.head(10)
```

### Добавляем столбцы во вспомогательные датасеты

Выполним все те же действия для датасетов с большими площадями и студиями.

```python
big_area_df['price_for_sq_m'] = (big_area_df['last_price'] / big_area_df['total_area']).round()
big_area_df['week_day_exposition'] = big_area_df['first_day_exposition'].dt.weekday
big_area_df['month_exposition'] = big_area_df['first_day_exposition'].dt.month
big_area_df['year_exposition'] = big_area_df['first_day_exposition'].dt.year
big_area_df['group_floors'] = big_area_df.apply(group_floors, axis=1)
big_area_df['cityCenters_nearest_km'] = (big_area_df['cityCenters_nearest'] / 1000).round()
big_area_df.head(10)
```

```python
studio_df['price_for_sq_m'] = (studio_df['last_price'] / studio_df['total_area']).round()
studio_df['week_day_exposition'] = studio_df['first_day_exposition'].dt.weekday
studio_df['month_exposition'] = studio_df['first_day_exposition'].dt.month
studio_df['year_exposition'] = studio_df['first_day_exposition'].dt.year
studio_df['group_floors'] = studio_df.apply(group_floors, axis=1)
studio_df['cityCenters_nearest_km'] = (studio_df['cityCenters_nearest'] / 1000).round()
studio_df.head(10)
```
